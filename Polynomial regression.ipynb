# Bangalore house price orediction

```python
import pandas as pd
import numpy as np
```

```python
#Import the data set
df=pd.read_csv('C:\\Users\\u\\Desktop\\Analysis data\\bangalore house price prediction OHE-data.csv')
df
```

```python
df.head()
```

```python
#define the dependent variable
# all column is x (independent feature) and only price is y (dependent feature)
x= df.drop('price', axis=1)
y= df['price']
print('shape of x =',x.shape)
print('shape of y =',y.shape)
```

```python
#Split the data into train and test with the help of sklearn library
#import train test split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 50)

print('shape of x_train = ', x_train.shape)
print('shape of y_train = ', y_train.shape)
print('shape of x_test = ', x_test.shape)
print('shape of y_test = ', y_test.shape)
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
sc.fit(x_train)
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

## Polynomial Linear regression, model training

```python
#Import LinearRegression and PolynomialFeatures from sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
#fit the data set and transform 
poly_reg = PolynomialFeatures(degree = 2)
poly_reg.fit(x_train)
x_train_poly = poly_reg.transform(x_train)
x_test_poly = poly_reg.transform(x_test)
```

```python
x_train_poly.shape, x_test_poly.shape
```

```python
# Fit the model in LinearRegression (Polynomial regression)

lr =LinearRegression()

lr.fit(x_train_poly, y_train)
```

```python
#Score of the model

lr.score(x_test_poly, y_test,)
```

```python
#Predict the value of first home (0th row and all column [x_test_poly[0,:])

lr.predict([x_test_poly[0,:]])
```

```python
#Predicted values are below from our model
y_pred = lr.predict(x_test_poly)
y_pred
```

```python
#Actual values are
y_test
```

```python
# import mean_squared_error from sklearn and find cost function(MSE, RMSE)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print('MSE is =', mse)
print('RMSE is =', rmse)
```
